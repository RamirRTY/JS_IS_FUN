```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
 x = sc.fit_transform(x)
```

```python
import numpy as np
```

```python
y = np.array(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.5 , random_state=42)
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units= 22 , activation="relu" , input_dim = 32))
ann.add(Dense(units= 22 , activation="relu" ))
ann.add(Dense(units= 22 , activation="relu" ))
ann.add(Dense(units= 22 , activation="relu" ))
ann.add(Dense(units=1 , activation= "sigmoid"))
```

```python
ann.compile(loss="binary_crossentropy" , optimizer="adam" , metrics=["accuracy"])
```

```python
ann.fit(x_train  , y_train , batch_size=32 , epochs=10)
```

```python
  y_new = ann.predict(x_test)
```

```python
Y_new = []

for element in y_new :
  if element > 0.5 :

   Y_new.append(1)
  else :
     Y_new.append(0)
```

```python
Y_new[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test , Y_new )
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix =  confusion_matrix(y_test , Y_new )
```

```python
print(matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
class_name = [0 , 1]
dataframe = pd.DataFrame(matrix , index = class_name , columns=class_name )
sns.heatmap(dataframe  , annot = True  , fmt = ".0f")
plt.ylabel("true class")
plt.xlabel("prediction class")
plt.show()
```
